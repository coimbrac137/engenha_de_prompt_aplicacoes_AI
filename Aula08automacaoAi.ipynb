#Temperatura
temperatura.png

```python
# Função que verifica a temperatura recebida como parâmetro
def verificar_temperatura(temp):

    # Verifica se a temperatura é maior que 30 graus
    if temp > 30:
        # Caso seja maior, exibe mensagem de alerta de superaquecimento
        print("ALERTA: Superaquecimento! Temperatura:", temp, "°C")
    else:
        # Caso contrário, informa que a temperatura está normal
        print("Temperatura normal:", temp, "°C")

# Solicita ao usuário que digite uma temperatura
temperatura = float(input("Digite a temperatura: "))

# Chama a função passando a temperatura informada
verificar_temperatura(temperatura)
```

#Automação de Arquivos
Automação.png

```python
import os
import shutil

# Define a pasta onde os arquivos serão criados e organizados
pasta = "arquivos_teste"

# Cria a pasta caso ela ainda não exista (evita erro)
os.makedirs(pasta, exist_ok=True)

# Lista de extensões que vamos usar para criar os arquivos
extensoes = [".pdf", ".jpg", ".xls", ".doc", ".py", ".ipynb"]

# ================================
# CRIAÇÃO DOS ARQUIVOS DE TESTE
# ================================

# Para cada extensão da lista...
for ext in extensoes:
    # Cria 10 arquivos com aquela extensão
    for i in range(10):
        # Monta o nome do arquivo (ex: arquivo_0.pdf)
        with open(f"{pasta}/arquivo_{i}{ext}", "w") as f:
            # Escreve um conteúdo simples dentro do arquivo
            f.write("teste")

print("Arquivos criados.\n")

# ================================
# ORGANIZAÇÃO DOS ARQUIVOS
# ================================

# Percorre todos os itens dentro da pasta
for arquivo in os.listdir(pasta):
    caminho = f"{pasta}/{arquivo}"  # caminho completo do arquivo

    # Verifica se é realmente um arquivo (e não uma pasta)
    if os.path.isfile(caminho):

        # Pega a extensão do arquivo (ex: pdf, jpg...)
        ext = arquivo.split(".")[-1]

        # Define a pasta de destino com base na extensão
        pasta_destino = f"{pasta}/{ext}"

        # Cria a pasta de destino se ainda não existir
        os.makedirs(pasta_destino, exist_ok=True)

        # Move o arquivo para a pasta correspondente
        shutil.move(caminho, f"{pasta_destino}/{arquivo}")

        # Mostra no terminal o que está acontecendo
        print(f"{arquivo} -> movido para pasta '{ext}'")

print("\nOrganização finalizada!")
```

#Consulta Simples a API do CEP
CEp.png

```python
import requests  # biblioteca para fazer requisições na internet

# ================================
# ENTRADA DO USUÁRIO
# ================================

# Pede o CEP para o usuário
cep = input("Digite o CEP: ")

# ================================
# CONSULTA NA API
# ================================

# Monta a URL da API do ViaCEP
url = f"https://viacep.com.br/ws/{cep}/json/"

try:
    # Faz a requisição para a API
    resposta = requests.get(url)

    # Converte a resposta para formato JSON (dicionário)
    dados = resposta.json()

    # ================================
    # TRATAMENTO DE ERRO (CEP INVÁLIDO)
    # ================================

    if "erro" in dados:
        print("CEP inválido!")

    else:
        # ================================
        # EXIBIÇÃO FORMATADA
        # ================================

        print("\nEndereço encontrado:")
        print("Rua:", dados["logradouro"])
        print("Bairro:", dados["bairro"])
        print("Cidade:", dados["localidade"])
        print("Estado:", dados["uf"])

except:
    # Caso aconteça erro de conexão ou outro problema
    print("Erro ao conectar com a API.")
```

#Sistema de Notificaçações
Notifi.png

```python
# Função para verificar se o valor é crítico
def verificar_condicao(valor):
    limite = 50  # Define o limite crítico
    return valor > limite  # Retorna True se for crítico

# Função para disparar o alerta
def disparar_alerta(valor):
    print("⚠️ ALERTA! Valor crítico atingido:", valor)  # Mostra alerta no console

# Função principal
def main():
    while True:  # Loop infinito para continuar rodando
        entrada = input("Digite um valor (ou 'sair' para encerrar): ")

        # Verifica se o usuário quer sair
        if entrada.lower() == "sair":
            print("Encerrando o programa...")
            break  # Sai do loop

        # Converte o valor para número
        valor = int(entrada)

        # Verifica a condição
        if verificar_condicao(valor):
            disparar_alerta(valor)  # Dispara alerta se for crítico
        else:
            print("Tudo normal 👍")  # Caso não seja crítico

# Executa o programa
main()
```

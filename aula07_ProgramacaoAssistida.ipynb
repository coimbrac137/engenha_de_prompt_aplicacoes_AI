Soma Lista


```python
# Cria uma lista com números inteiros
numeros = [1, 3, 6, 7, 8]

# Define uma variável para guardar o total da soma
soma = 0

# Percorre cada valor dentro da lista 'numeros'
for num in numeros:
  # Soma o valor atual à variável 'soma'
  soma += num

# Exibe o resultado final da soma de todos os valores
print(soma)
```
